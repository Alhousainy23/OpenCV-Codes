```python
import os 
import cv2 as cv 
import dlib 
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.spatial import distance 
from imutils import face_utils 
import imutils 
import datetime 
import time 
import playsound 
count = 0 
alarm_on = False 
os.chdir(R"D:\\AI\\AI Models\\3. OpenCV")  
model_path = "shape_predictor_68_face_landmarks.dat"
alarm_path = "assets_alarm.mp3"
#-------------------------------------------------------------- 
def eye_aspect_ratio(eye):
    A = distance.euclidean(eye[1],eye[5])
    B = distance.euclidean(eye[2],eye[4])
    C = distance.euclidean(eye[0],eye[3])
    ear = (A+B)/(2.0 * C) #ear ==> Eye Aspect Ratio 
    return ear 
#-----------------------------------------------------------------
def  detect():
    thresh = 0.25 
    frame_check = 10 # default = 20 
    detect = dlib.get_frontal_face_detector()
    predict= dlib.shape_predictor(model_path)
    (lstart,lend) = face_utils.FACIAL_LANDMARKS_68_IDXS['left_eye']
    (rstart,rend) = face_utils.FACIAL_LANDMARKS_68_IDXS['right_eye']
    cap = cv.VideoCapture(0)
    flag = 0 
    while True: 
        ret,frame = cap.read()
        frame = imutils.resize(frame,width=640,height=640)
        gray = cv.cvtColor(frame,cv.COLOR_BGR2GRAY)
        subjects = detect(gray,0)
        for subject in subjects:
            shape = predict(gray,subject)
            shape = face_utils.shape_to_np(shape)#Convert to numpy array 
            leftEye = shape[lstart:lend] 
            print("The Left Eye Is --> ",leftEye)
            rightEye = shape[rstart:rend]
            print("The Right Eye Is --> ",rightEye)
            leftEAR = eye_aspect_ratio(leftEye)
            rightEAR= eye_aspect_ratio(rightEye)
            ear = (leftEAR + rightEAR)/2.0
            leftEyeHull = cv.convexHull(leftEye)
            rightEyeHull= cv.convexHull(rightEye)
            cv.drawContours(frame,[leftEyeHull],-1,(0,255,0),1)
            cv.drawContours(frame,[rightEyeHull],-1,(0,255,0),1)
            if ear < thresh:         
                flag += 1
                print("Eye Aspect Ratio is ==> ", flag)
                if flag >= frame_check:
                    if not alarm_on:
                        alarm_on = True
                        time.sleep(0.01)
                    datet = str(datetime.datetime.now())
                    cv.putText(frame, "****************Drowsy!****************", (10, 30),cv.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
                    #stop = timeit.default_timer()
                    cv.putText(frame,'sleeping time 00:'+str(flag)+'sec', (10, 60),cv.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
                    print("Drowsy")
                    playsound.playsound(alarm_path, True)   
            else:
                flag = 0
                alarm_on = False
        cv.imshow("Frame", frame)
        key = cv.waitKey(1) & 0xFF
        if key == ord("q"):
            break
    cv.destroyAllWindows()
    cap.release()
if __name__ == '__main__':
    detect()
#Footer
```
