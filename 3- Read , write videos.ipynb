# **1. Read and Write and Show Video from camera by using OpenCv**

```python
import cv2 
cap = cv2.VideoCapture(0)#Because to rrecord video from camera 
fourcc = cv2.VideoWriter_fourcc(*'XVID') # because tunning resolution (Pixels , colors )
#Can write fourcc another way 
# fourcc = cv2.VideoWriter_fourcc('X','V','I','D')
out = cv2.VideoWriter('Video_Capture.avi',fourcc,20.0,(640,480))
#Video_Capture.avi --> Video name on device 
#20.0 --> Number of frames each second 
#(40,480) --> Height and width for each frame 
print(cap.isOpened())#Check True Or False 
#While (True) --> Right way and can used it 
while (cap.isOpened()):
    ret,frame = cap.read()#frame --> because save frames comes from video  , ret = True or false 
    if ret ==True : 
        print(cap.get(cv2.CAP_PROP_FRAME_WIDTH))#Read width from coming frame 
        print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))#Read height from comming frame
        out.write(frame)#Save Frame On device
        gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY) # Convert video to gray 
        cv2.imshow('frame',gray)
        if cv2.waitKey(1) & 0xff==ord('q'): break 
    else : break
cap.release()
out.release()
cv2.destroyAllWindows()
'''
import cv2 
capture = cv2.VideoCapture(0)
fourcc = cv2.VideoWriter_fourcc(*'XVID')
output = cv2.VideoWriter('Movie.avi',fourcc,20.0 , (640,480))
print(capture.isOpened())
while(capture.isOpened()):
    ret , frame = capture.read()
    if ret == True:
        print(capture.get(cv2.CAP_PROP_FRAME_WIDTH))
        print(capture.get(cv2.CAP_PROP_FRAME_HEIGHT))
        output.write(frame)
        gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        cv2.imshow('frame',gray)
        if cv2.waitKey(1) & 0xff==ord('q'): break 
    else : break 
capture.release()
output.release()
cv2.destroyAllWindows()
'''
```

```python
import cv2

# فتح الكاميرا
cap = cv2.VideoCapture(0)

# التحقق من استعداد الكاميرا
if not cap.isOpened():
    print("لا يمكن فتح الكاميرا")
    exit()

while True:
    # قراءة الإطار الحالي من الكاميرا
    ret, frame = cap.read()

    # التحقق من نجاح قراءة الإطار
    if not ret:
        print("لا يمكن قراءة الإطار من الكاميرا")
        break

    # عرض الإطار المقروء
    cv2.imshow('Camera', frame)

    # الانتظار لضغط مفتاح 'q' لإنهاء البرنامج
    if cv2.waitKey(1) == ord('q'):
        break

# إغلاق الكاميرا وإغلاق النوافذ
cap.release()
cv2.destroyAllWindows()
```
