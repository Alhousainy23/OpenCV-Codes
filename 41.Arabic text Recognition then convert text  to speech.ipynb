```python
def download(url, file_name):
    # open in binary mode
    with open(file_name, "wb") as file:
        # get request
        response = get(url)
        # write to file
        file.write(response.content)

download("https://github.com/tesseract-ocr/tessdata/blob/main/ara.traineddata","ara.traineddata")
```

```python
import cv2 as cv 
image_cv = cv.imread("Arabic_OCR.png")
plt.figure(figsize=(20,20))
plt.imshow(image_cv,cmap='gray')
plt.title('Template'),plt.xticks([]),plt.yticks([])
plt.show()
```

```python
arabix_text = pt.image_to_string(image_cv,lang='ara',config=".")
print(arabix_text)
```
