* **Some Notes**
1. detect distance = 9.144 meter ==> divide on `1000` to convert to KM 
2. detect time = time and divide on `3600` to convert Hours
-----------------------------------------------------------------
1.  Read Video - Create Count - Calculate Start Time by using `Time Func` 
2.  Create cascade classifier and using `(cars1.xml)` --> because trining for cars
---------------------------------------------------------------------------------
* Create While Loop 
1. Check if `image_type` is `null` --> make `break` ==> meaning the video is not contain any frames
2. If `True` --> Take Video frames and abbply blur filter because making `Smoothing` for image
3. Make detection by using `car_cascade` , and using `detectMultiScale`
4. Draw 2 lines (line 1 , line 2)
5. Create for Loop to draw `Rectangle` for each Cars
6. draw `Circle` enter for each cars
7. Create To Loops --> `Loop 1` take time when `(line 1 matching with point)`, `Loop2` take end time `Time(line2) - T1` if not matching in `loop1`
-------------------------------------------------------------------------------------------------

```python
import cv2 as cv 
import numpy as np
import matplotlib.pyplot as plt 
import os 
import time 
os.chdir(R"D:\\AI\\AI Models\\3. OpenCV\\0- OpenCV\\Data\\Videos")
video_src = "video3.mp4"
cascade_src = R"D:\\AI\\AI Models\\3. OpenCV\\0- OpenCV\\Data\\calculte-the-vehicle-speed-using-open-CV-Computer-vision-main\\cars1.xml"
#Detect Line 1 (a)
ax1,ay,ax2 = 70,90,230 
#Detect Line 2(b)
bx1,by,bx2 = 15,125,225
#Create Function To calaculate The Speed 
def speed_cell(time):
    try: 
        speed = (9.144/1000)/(time/3600)
        return speed 
    except ZeroDivisionError: print("Can Not Devide by Zero!")
#--------------------------------------------------------------------
#Car num 
i = 1
start_time = time.time()
#Read Video 
cap = cv.VideoCapture(video_src)
car_cascade = cv.CascadeClassifier(cascade_src)
#-------------------------------------------------------------------------
while True:
    #-------------Read Video-------------------------------------------------------
    ret,image =cap.read()
    #Can used if ret == false : break --> but using another way 
    if (type(image) == type(None)):break 
    #----------------Blurred to have extracter detection--------------------------- 
    blurred = cv.blur(image,ksize=(15,15))
    gray = cv.cvtColor(blurred,cv.COLOR_BGR2GRAY)
    cars = car_cascade.detectMultiScale(gray,1.1,2)
    #--------------------Draw 2 Lines (Line 1, Line 2)------------------------------
    #1. Line 1 
    cv.line(image,(ax1,ay),(ax2,ay),(255,0,0),2)
    #2. Line 2
    cv.line(image,(bx1,by),(bx2,by),(255,0,0),2)
    #---------------Create For Loop To dreaw rectangle on car-----------------------
    for (x,y,w,h) in cars:
        cv.rectangle(image,(x,y),(x+w,y+h),(0,0,255),2)
        cv.circle(image,(int((x+x+w)/2),int((y+y+h)/2)),3,(0,255,0),-1)
        while int(ay) == int((y+y+h)/2):
            cv.line(image,(ax1,ay),(ax2,ay),(0,255,0),2)
            start_time = time.time()
            break
        while int (ay) <= int((y+y+h)/2):
            if int(by) <= int((y+y+h)/2) & int( by + 10 ) >= int(( y + y+h)/2):   
             cv.line(image,(bx1,by),(bx2,by),(0,255,0),2)
             speed = speed_cell(time.time() - start_time)
             print("# Car Number " + str(i) + " Speed Is --> "+str(speed)+ "  KM/H")
             i+=1
             break
            else :
                cv.putText(image, "Calcuting", (100,200), cv.FONT_HERSHEY_SIMPLEX,1,(0,255,0),3)
                break    
    cv.imshow('video', image) 
    if cv.waitKey(33) == 27:
        break  
cap.release()    
cv.destroyAllWindows()
#cv2.imshow('video', img) 
#if cv2.waitKey(33) == 27:break
#cap.release()#close cam
#cv2.destroyAllWindows() #close window
```
