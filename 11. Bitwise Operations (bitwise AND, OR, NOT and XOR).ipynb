**Create Images**

```python
import cv2 as cv 
import numpy as np 
img1 = np.zeros((250,500,3),np.uint8)
img1 = cv.rectangle(img1, (200,0),(300,100),(255,255,255),-1)
img2 = np.full((250,500,3),255, dtype=np.uint8)
img2 = cv.rectangle(img2,(0,0),(250,250),(0,0,0),-1) 
cv.imshow('Image 1',img1)
cv.imshow('Image2',img2)
cv.waitKey(0)
cv.destroyAllWindows()
```

**Applied Bitwise**

```python
bit_and = cv.bitwise_and(img2,img1)
bit_or = cv.bitwise_or(img2,img1)
bit_xor = cv.bitwise_xor(img2,img1)
bit_not1 =cv.bitwise_not(img1)
bit_not2 =cv.bitwise_not(img2)
bit_and = cv.resize(bit_and,(250,250))
bit_or  = cv.resize(bit_or,(250,250))
bit_xor = cv.resize(bit_xor,(250,250))
bit_not1= cv.resize(bit_not1,(250,250))
bit_not2= cv.resize(bit_not2,(250,250)) 
cv.imshow('Bitwise And ',bit_and)
cv.imshow('Bitwise Or ',bit_or)
cv.imshow('Bitwise XOR',bit_xor)
cv.imshow('Bitwise NOT Img1',bit_not1)
cv.imshow('Bitwise NOT Img2',bit_not2)
cv.waitKey(0)
cv.destroyAllWindows()
```

* Trackbar --> هو عبارة عن شريط يقوم بتوضيح اللالوان والتحكم فيها ، والقيام بعملية تغير لها
* هنقوم بإنشاء 4 من TrackBar 
* Trackbar --> (B ,G ,R , Switch)
* `Trackbar` --> Take Four Arguments 
* **`trackbarname` , `windowname` , `value` , `count`,  `Onchange`**
* cv.createTrackbar('B','Image' , `Value` 0 , `Count` 255) 
  * رقمين 0 و 255 معنهاهم انه هيبدا من 0 لحد 255 
* `Switch Trackbar in code` --> 0 `OFF` & 1 `ON`
* `Trackbarpos` --> بترجع لي او بتجيب لي القيمة بتاعت اللون او الرقم اللي واقف عليه دلوقتي 

```python
import numpy as np 
import cv2 as cv 
def nothing(x): print(x)
#Create a black image , window
img = np.zeros((300,512,3),dtype = np.uint8)
img = cv.resize(img,(400,300))
#To Create Name For Window
cv.namedWindow('Image')
#Create Trackbar 
cv.createTrackbar('B','Image' , 0 , 255,nothing)
cv.createTrackbar('G','Image' , 0 , 255,nothing)
cv.createTrackbar('R','Image' , 0 , 255,nothing)
switch = '0 : OFF\n 1 : ON'
cv.createTrackbar(switch,'Image',0,1,nothing)
while(1):
    cv.imshow('Image',img)
    k = cv.waitKey(1)
    if k == 27 : break 
    b = cv.getTrackbarPos('B','Image')
    g = cv.getTrackbarPos('G','Image')
    r = cv.getTrackbarPos('R','Image')
    s = cv.getTrackbarPos(switch , 'Image')
    if s == 0 : img[:] = 0
    else : img[:] = [b,g,r]
cv.destroyAllWindows()
```
