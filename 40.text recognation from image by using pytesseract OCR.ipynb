* **`text and number detection` by using pytesseract OCR**

```python
import cv2 as cv 
from pytesseract import image_to_string , image_to_boxes , image_to_data , pytesseract 
import os 
os.chdir(R"D:\\AI\\AI Models\\3. OpenCV\\0- OpenCV\\Data\\Images")
pytesseract.tesseract_cmd = r"C:\Users\Alhou\AppData\Local\Programs\Tesseract-OCR\tesseract.exe" 
image = cv.imread("testt.png")
Himg,Wimg,_ = image.shape
boxes = image_to_boxes(image,lang="eng")
print(boxes)
for b in boxes.splitlines():
    print(b)
    b = b.split(" ")
    print(b)
    x,y,w,h = int(b[1]),int(b[2]),int(b[3]),int(b[4]) 
    #cv.rectangle(image,(x,y),(x+w,y+h),(0,255,0),3) --> I Need Change Coordinates 
    cv.rectangle(image, (x,Himg - y), (w,Himg-h),(0,255,0),3)
    cv.putText(image,b[0],(x,Himg-y+25),cv.FONT_HERSHEY_COMPLEX,1,(0,0,255),1)
cv.imshow("Image",image)
#cv.imshow("Adaptive Thresh",adaptive_thresh)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import cv2 as cv 
from pytesseract import image_to_string , image_to_boxes , image_to_data , pytesseract 
import os 
os.chdir(R"D:\\AI\\AI Models\\3. OpenCV\\0- OpenCV\\Data\\Images")
pytesseract.tesseract_cmd = r"C:\Users\Alhou\AppData\Local\Programs\Tesseract-OCR\tesseract.exe" 
image = cv.imread("test.png")
Himg,Wimg,_ = image.shape
boxes =image_to_data(image,lang="eng")
print(boxes)
for x , b in enumerate(boxes.splitlines()):
    if x !=0:
        b = b.split()
        print(b)
        if len(b) == 12 :
            x,y,w,h = int(b[6]),int(b[7]),int(b[8]),int(b[9]) 
            #cv.rectangle(image,(x,y),(x+w,y+h),(0,255,0),3) --> I Need Change Coordinates 
            cv.rectangle(image, (x,y), (x+w,y+h),(0,255,0),3)
            cv.putText(image,b[11],(x,y),cv.FONT_HERSHEY_COMPLEX,1,(0,0,255),2)
cv.imshow("Image",image)
#cv.imshow("Adaptive Thresh",adaptive_thresh)
cv.waitKey(0)
cv.destroyAllWindows()
```
