```python
import cv2 as cv 
import numpy as np 
import matplotlib.pyplot as plt 
import os 
plt.style.use('ggplot')
os.chdir(R"D:\\AI\\AI Models\\3. OpenCV\\0- OpenCV\\Data\\Images")
img = cv.imread("noise.jpeg",0)
#1. Apply Global Thresholding --> هنا أنا عارف القيمة بتاعت ثيرشولد 
ret1,th1 = cv.threshold(img,127,255,cv.THRESH_BINARY)
#2. Apply Outsu's Thresholding --> هنا انا مش عارف القيمة فخليتها 0 عكس الول انا عارفه وهو 127 
ret2,th2 = cv.threshold(img,0,255,cv.THRESH_BINARY + cv.THRESH_OTSU)
# 3. Otsu's thresholding after Gaussian filtering 
blur = cv.GaussianBlur(img,(5,5),0)
ret3,th3 = cv.threshold(blur,0,255, cv.THRESH_BINARY + cv.THRESH_OTSU)
#4. Apply plot all the images and thier histograms
images = [img, 0, th1,img, 0, th2, blur, 0, th3]
titles = ['Original Noisy Image','Histogram','Global Thresholding (v=127)','Original Noisy Image','Histogram',"Otsu's Thresholding",'Gaussian filtered Image','Histogram',"Otsu's Thresholding"]
for i in range(3):
    plt.subplot(3,3,i*3+1),plt.imshow(images[i*3],'gray')
    plt.title(titles[i*3]), plt.xticks([]), plt.yticks([])
    plt.subplot(3,3,i*3+2),plt.hist(images[i*3].ravel(),256)
    plt.title(titles[i*3+1]), plt.xticks([]), plt.yticks([])
    plt.subplot(3,3,i*3+3),plt.imshow(images[i*3+2],'gray')
    plt.title(titles[i*3+2]), plt.xticks([]), plt.yticks([])
plt.show()
```
